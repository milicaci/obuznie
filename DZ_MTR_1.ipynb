```python
# подсет выручки
def doxod(a):
    return a['Quantity'] * a['UnitPrice'] # подсет выручки
df['Doxod'] = df.apply(doxod, axis=1)
df.head()
```

```python
df['Doxod'].sum()
```

```python
df[['CustomerID', 'Country', 'Doxod']].groupby('CustomerID').sum().sort_values(by='Doxod', ascending=False).reset_index()
```
