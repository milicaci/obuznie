```python
import pandas as pd
import numpy as np
```

Задание 1. Загрузка данных

```python
df = pd.read_csv('horse_data.csv', na_values = '?', header = None)
df
```

```python
### переименуем столбцы
df.rename(columns = {0:'surgery', 1:'Age', 3:'rectal temperature', 4:'pulse', 5: 'respiratory rate', 6:'temperature of extremities', 10:'pain', 22:'outcome'}, inplace=True)
```

```python
df_kor = df[['surgery','Age','rectal temperature','pulse','respiratory rate','temperature of extremities','pain', 'outcome']]
df_kor.head()
```

```python
Задание 2. Первичное изучение данных
```

```python
# стандартное отклонение набора данных
df_kor.describe()
```

```python
# выбросы через межквартльный размах (на примере роста)
q1 = df_kor['respiratory rate'].quantile(0.25)
q3 = df_kor['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_kor[df_kor['respiratory rate'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
```

При высоком пульсе частота дыхания повышенная, но есть  исключения.

```python
pd.concat([df_kor, remove_outliers]).drop_duplicates(keep=False)
```
